```python
import matplotlib as mlp
import matplotlib.pyplot as plt
plt.style.use('seaborn')
mlp.rcParams['font.family'] = 'serif'
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
filename = "py4fi/jupyter36/source/tr_eikon_eod_data.csv"
f = open(filename, 'r')
df = pd.read_csv(filename, index_col=0, parse_dates=True)
df.plot(figsize=(12, 10), subplots=True)
```

```python
instruments = ['Apple Stock', 'Microsoft Stock',
                       'Intel Stock', 'Amazon Stock', 'Goldman Sachs Stock',
                       'SPDR S&P 500 ETF Trust', 'S&P 500 Index',
                       'VIX Volatility Index', 'EUR/USD Exchange Rate',
                       'Gold Price', 'VanEck Vectors Gold Miners ETF',
                       'SPDR Gold Trust']
for ric, name in zip(df.columns, instruments):
    print('%8s | %s' % (ric, name))
```

```python
sym = 'AAPL.O'

aapl = pd.DataFrame(df[sym]).dropna()
aapl.tail()
```

```python
window = 20
df['min'] = df[sym].rolling(window=window).min()
df.dropna().head()
```

```python
df['SMA 1'] = df[sym].rolling(window=42).mean()
df['SMA 2'] = df[sym].rolling(window=242).mean()
df[[sym, 'SMA 1', 'SMA 2']].dropna().head()
```

```python
df[[sym, 'SMA 1', 'SMA 2']].dropna().plot(figsize=(12, 10))
```

```python
df.dropna(inplace=True)
df['positions'] = np.where(df['SMA 1'] > df['SMA 2'], 1, -1)
ax = df[[sym, 'SMA 1', 'SMA 2', 'positions']].plot(figsize=(12, 10), secondary_y = 'positions')
#ax.get_legend().set_bbox_to_anchor((0.25, 0.85))
```
