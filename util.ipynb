```python
import pandas as pd 
```

```python
def import_data(symbols):
    start = '2014-12-22'
    end = '2019-12-22'
    dates = pd.date_range(start, end)
    df = pd.DataFrame(index=dates)
    for symbol in symbols:
        temp = pd.read_csv('data/%s.csv' % symbol, index_col='Date', usecols=['Date', 'Adj Close'], na_values=['nan'], parse_dates=True)
        temp = temp.rename(columns={'Adj Close' : symbol})
        df= df.join(temp)
    return df
        
```

```python
df = import_data(['XOM', 'TSLA'])
df.dropna().head()
```

```python
df = df / df.iloc[0, :]
```
