```python
import pandas as pd
```

```python
def get_max_closed(symbol):
    df = pd.read_csv('data/%s.csv' % symbol)
    return df['Adj Close'].max()
```

```python
symbols = ['TSLA', 'XOM', 'GOOGL']
for symbol in symbols:
    print("Max close")
    print(symbol, get_max_closed(symbol))
```
